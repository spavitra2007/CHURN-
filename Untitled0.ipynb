```python
import pandas as pd

cols = [
    "total_complaints",
    "complaint_resolution_time",
    "unresolved_complaint_count",
    "branch_visit_count",
    "call_center_interaction_count",
    "relationship_manager_interaction_count",
    "service_request_count",
    "escalation_count",
    "campaign_received_count",
    "campaign_response_count",
    "cross_sell_offer_count",
    "upsell_offer_count",
    "retention_offer_received",
    "retention_offer_accepted",
    "discount_or_fee_waiver_received"
]

# 1. Summary stats table
summary = df[cols].describe().T

# 2. Missing values
missing = df[cols].isnull().sum().rename("missing_count")

# 3. Combine into one table
final_table = summary.join(missing)

# reorder columns nicely
final_table = final_table[[
    "count", "missing_count", "mean", "std", "min", "25%", "50%", "75%", "max"
]]

final_table
```

```python
import pandas as pd

cols_risk = [
    "total_complaints",
    "unresolved_complaint_count",
    "escalation_count",
    "call_center_interaction_count",
    "service_request_count",
    "branch_visit_count"
]

cols_positive = [
    "retention_offer_accepted",
    "campaign_response_count",
    "cross_sell_offer_count",
    "upsell_offer_count"
]

# Create scores
df["service_risk_score"] = df[cols_risk].sum(axis=1)
df["service_engagement_score"] = df[cols_positive].sum(axis=1)
df["churn_pressure_score"] = df["service_risk_score"] - df["service_engagement_score"]

# Final table
final_table = df[[
    "service_risk_score",
    "service_engagement_score",
    "churn_pressure_score"
]]

final_table
```
